```python
import pandas as pd
import numpy as np
import json
```

```python
df_races = pd.read_json(r'C:\Users\aalbe\OneDrive\Escritorio\PI01_DATA02\Datasets/races.json')
```

```python
df_races
```

```python
# hacemos este dataframe para reducir el dataset que estamos manejando
contador_carreras = pd.DataFrame(df_races['year'].value_counts())
```

```python
contador_carreras
```

```python
# reseteamos el indice para que el año pase a ser una columna
contador_carreras.reset_index(inplace=True)
```

```python
# cambiamos de nombre las columnas
contador_carreras.rename({'year': 'repeticiones'}, axis=1, inplace=True)
```

```python
contador_carreras.rename({'index': 'year'}, axis=1, inplace=True)
```

```python
contador_carreras
```

```python
# máscara que muestra mayores repeticiones
mayor_repeticion = contador_carreras['repeticiones'] == contador_carreras['repeticiones'].max()
```

```python
# filtramos el año de mayor repetición
int(contador_carreras[mayor_repeticion]['year'].to_string(index = False))
```

```python
# entonces queda:
print('el año con mayor repeticiones es', int(contador_carreras[mayor_repeticion]['year'].to_string(index = False)), 'con un total de', contador_carreras['repeticiones'].max(), 'repeticiones')
```
